Problema 3: La resta inversaPlanteamiento:Buscamos un número $x$ tal que, al quitarle 7, nos queden 18.Paso 1: Resolución MatemáticaDefinir la variable: Sea $x$ el número buscado.Plantear la ecuación: $x - 7 = 18$Despejar la variable: Para dejar sola a la $x$, el 7 que está restando pasa al otro lado sumando (operación inversa):$$x = 18 + 7$$Resultado:$$x = 25$$

```python
# Problema 3: x - 7 = 18
resultado_resta = 18
valor_restado = 7

# Despejamos sumando: x = 18 + 7
x = resultado_resta + valor_restado

print(f"El número buscado es: {x}")
```

Problema 4: Operaciones combinadas (Multiplicación y Suma)Planteamiento:Aquí tenemos dos pasos: el "triple" de un número ($3x$) y luego sumarle 4.Paso 1: Resolución MatemáticaDefinir la variable: Sea $x$ el número.Plantear la ecuación: $3x + 4 = 31$Despejar la variable:Primero quitamos lo que suma: El 4 pasa restando.$$3x = 31 - 4$$$$3x = 27$$Luego quitamos lo que multiplica: El 3 pasa dividiendo.$$x = \frac{27}{3}$$Resultado:$$x = 9$$

```python
# Problema 4: 3x + 4 = 31
total = 31
suma = 4
multiplicador = 3

# Primero restamos el valor que suma
paso1 = total - suma
# Luego dividimos por el multiplicador
x = paso1 / multiplicador

print(f"El número buscado es: {int(x)}")
```

Problema 5: Dimensiones de un terreno (Geometría)Planteamiento:Un rectángulo tiene dos anchos y dos largos.Ancho: $x$Largo: $x + 5$Perímetro: Suma de todos los lados = 50.Paso 1: Resolución MatemáticaDefinir la variable:Ancho = $x$Largo = $x + 5$Plantear la ecuación (Perímetro): $Ancho + Ancho + Largo + Largo = 50$$$x + x + (x + 5) + (x + 5) = 50$$Simplificar: Sumamos las $x$ y los números.$$4x + 10 = 50$$Despejar la variable:Restamos 10: $4x = 50 - 10 \rightarrow 4x = 40$Dividimos entre 4: $x = \frac{40}{4}$Resultado:Ancho ($x$): 10 metros.Largo ($x + 5$): 15 metros.

```python
# Problema 5: Perímetro de un rectángulo (4x + 10 = 50)
perimetro_total = 50
diferencia_largo = 5 # El largo mide 5 más que el ancho

# Según la ecuación 4x + (2 * diferencia) = Perímetro
# Despejamos x (el ancho)
ancho = (perimetro_total - (2 * diferencia_largo)) / 4
largo = ancho + diferencia_largo

print(f"El ancho mide: {int(ancho)}m")
print(f"El largo mide: {int(largo)}m")
```
