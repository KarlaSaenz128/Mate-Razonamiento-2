* DIseñar un programa que permita administrar una lista de medico dentista


```python
medicos = []
while True :
  print("MENU DE MEDICO DENTISTA")
  print("1. Agregar medico dentista")
  print("2. cambiar medico dentista")
  print("3. Eliminar medico dentista")
  print("4.ver lista de medico dentista")
  print("5. salir")
  opcion = input("ingrese una opcion: ")
  if opcion == "1":
    nombre = input("ingrese el nombre del medico: ")
    medicos.append(nombre)
    print("medico agregado ")
  elif opcion == "2":
    nombre = input ("ingrese el nombre del medico a cambiar : ")
    nuevo_nombre = input(" ingrese el nuevo nombre del medico : ")
    if nombre in medicos:
      indice = medicos.index(nombre)
      medicos[indice] = nuevo_nombre
      print("medico cambiado")
    else :
      print("medico no encontrado")
  elif opcion == "3":
    nombre = input("ingrese el nombre del medico a eliminar :")
    if nombre in medicos :
      medicos.remove(nombre)
      print("medico eliminado")
    else:
      print("medico no encontrado")
  elif opcion == "4":
    if medicos:
      print("Lista de medicos:")
      for medico in medicos:
        print(f"- {medico}")
    else:
      print("No hay medicos en la lista.")
  elif opcion == "5":
    print("Saliendo del programa.")
    break
  else:
    print("Opcion invalida. Por favor, ingrese un numero del 1 al 5.")
```
