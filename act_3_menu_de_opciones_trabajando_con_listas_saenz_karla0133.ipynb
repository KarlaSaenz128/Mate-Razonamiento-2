* Diseñar un programa que permita manejar una lista de frutas de temporada.


```python
frutas_temporada = []

while True:
    print("\n--- SISTEMA DE GESTIÓN DE FRUTAS DE TEMPORADA ---")
    print("1. Agregar fruta")
    print("2. Modificar fruta")
    print("3. Eliminar fruta")
    print("4. Ver inventario de frutas")
    print("5. Salir")

    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        nombre_fruta = input("Ingrese el nombre de la fruta que desea agregar: ")
        frutas_temporada.append(nombre_fruta)
        print(f"'{nombre_fruta}' ha sido añadida con éxito.")

    elif opcion == "2":
        buscar_fruta = input("Ingrese el nombre de la fruta que desea cambiar: ")
        if buscar_fruta in frutas_temporada:
            nueva_fruta = input("Ingrese el nuevo nombre de la fruta: ")
            indice = frutas_temporada.index(buscar_fruta)
            frutas_temporada[indice] = nueva_fruta
            print("Registro actualizado correctamente.")
        else:
            print("Error: La fruta no se encuentra en la lista.")

    elif opcion == "3":
        eliminar_fruta = input("Ingrese el nombre de la fruta a eliminar: ")
        if eliminar_fruta in frutas_temporada:
            frutas_temporada.remove(eliminar_fruta)
            print(f"'{eliminar_fruta}' ha sido eliminada.")
        else:
            print("Error: Fruta no encontrada.")

    elif opcion == "4":
        if frutas_temporada:
            print("\n--- FRUTAS DISPONIBLES EN TEMPORADA ---")
            # Usamos enumerate para que la lista se vea numerada
            for i, fruta in enumerate(frutas_temporada, 1):
                print(f"{i}. {fruta}")
        else:
            print("La lista de temporada está vacía actualmente.")

    elif opcion == "5":
        print("Cerrando el sistema de frutas. ¡Hasta luego!")
        break

    else:
        print("Opción no válida. Intente con un número del 1 al 5.")
```
